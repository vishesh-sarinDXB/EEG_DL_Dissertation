```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_real = pd.read_csv('../summary/mi_real/all.csv')
```

```python
all_real['x'] = 'all'
```

```python
all_real['hue'] = 'mi_real'
```

```python
all_cv = pd.read_csv('../summary/mi_cv/all/all.csv')
```

```python
all_cv['x'] = 'all'
```

```python
all_cv['hue'] = 'mi_cv'
```

```python
test = all_real.append(all_cv).reset_index().drop(columns = 'index')
```

```python
plt.figure(figsize=(15,10))
sns.violinplot(x="x", y="test_accuracy", hue="hue", data=test, split=True)
```
