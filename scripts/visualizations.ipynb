```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_real = pd.read_csv('../summary/mi_real/all.csv')
all_cv = pd.read_csv('../summary/mi_cv/all/all.csv')
all_str = 'all'

all_channels_alphamu_real = pd.read_csv('../summary/mi_real/all_channels_alphamu.csv')
all_channels_alphamu_cv = pd.read_csv('../summary/mi_cv/all_channels_alphamu/all_channels_alphamu.csv')
all_channels_alphamu = 'all_channels_alphamu'

bio_channelsConly_alphamu_real = pd.read_csv('../summary/mi_real/bio_channelsConly_alphamu.csv')
bio_channelsConly_alphamu_cv = pd.read_csv('../summary/mi_cv/bio_channelsConly_alphamu/bio_channelsConly_alphamu.csv')
bio_channelsConly_alphamu = 'bio_channelsConly_alphamu'

bio_channelsOConly_alphamu_real = pd.read_csv('../summary/mi_real/bio_channelsOConly_alphamu.csv')
bio_channelsOConly_alphamu_cv = pd.read_csv('../summary/mi_cv/bio_channelsOConly_alphamu/bio_channelsOConly_alphamu.csv')
bio_channelsOConly_alphamu = 'bio_channelsOConly_alphamu'

bio_channelsOonly_alphamu_real = pd.read_csv('../summary/mi_real/bio_channelsOonly_alphamu.csv')
bio_channelsOonly_alphamu_cv = pd.read_csv('../summary/mi_cv/bio_channelsOonly_alphamu/bio_channelsOonly_alphamu.csv')
bio_channelsOonly_alphamu = 'bio_channelsOonly_alphamu'

bio_channelsOPT_real = pd.read_csv('../summary/mi_real/bio_channelsOPT.csv')
bio_channelsOPT_cv = pd.read_csv('../summary/mi_cv/bio_channelsOPT/bio_channelsOPT.csv')
bio_channelsOPT = 'bio_channelsOPT'

bio_channelsOPT_alphamu_real = pd.read_csv('../summary/mi_real/bio_channelsOPT_alphamu.csv')
bio_channelsOPT_alphamu_cv = pd.read_csv('../summary/mi_cv/bio_channelsOPT_alphamu/bio_channelsOPT_alphamu.csv')
bio_channelsOPT_alphamu = 'bio_channelsOPT_alphamu'

bio_channelsPT_real = pd.read_csv('../summary/mi_real/bio_channelsPT.csv')
bio_channelsPT_cv = pd.read_csv('../summary/mi_cv/bio_channelsPT/bio_channelsPT.csv')
bio_channelsPT = 'bio_channelsPT'

bio_channelsPT_alphamu_real = pd.read_csv('../summary/mi_real/bio_channelsPT_alphamu.csv')
bio_channelsPT_alphamu_cv = pd.read_csv('../summary/mi_cv/bio_channelsPT_alphamu/bio_channelsPT_alphamu.csv')
bio_channelsPT_alphamu = 'bio_channelsPT_alphamu_cv'
```

```python
experiment = 'experiment'
data = 'data'
mi_real = 'mi_real'
mi_cv = 'mi_cv'
```

```python
all_real[experiment] = all_str
all_real[data] = mi_real

all_cv[experiment] = all_str
all_cv[data] = mi_cv

all_channels_alphamu_real[experiment] = all_channels_alphamu
all_channels_alphamu_real[data] = mi_real

all_channels_alphamu_cv[experiment] = all_channels_alphamu
all_channels_alphamu_cv[data] = mi_cv

bio_channelsConly_alphamu_real[experiment] = bio_channelsConly_alphamu
bio_channelsConly_alphamu_real[data] = mi_real

bio_channelsConly_alphamu_cv[experiment] = bio_channelsConly_alphamu
bio_channelsConly_alphamu_cv[data] = mi_cv

bio_channelsOConly_alphamu_real[experiment] = bio_channelsOConly_alphamu
bio_channelsOConly_alphamu_real[data] = mi_real

bio_channelsOConly_alphamu_cv[experiment] = bio_channelsOConly_alphamu
bio_channelsOConly_alphamu_cv[data] = mi_cv

bio_channelsOonly_alphamu_real[experiment] = bio_channelsOonly_alphamu
bio_channelsOonly_alphamu_real[data] = mi_real

bio_channelsOonly_alphamu_cv[experiment] = bio_channelsOonly_alphamu
bio_channelsOonly_alphamu_cv[data] = mi_cv

bio_channelsOPT_real[experiment] = bio_channelsOPT
bio_channelsOPT_real[data] = mi_real

bio_channelsOPT_cv[experiment] = bio_channelsOPT
bio_channelsOPT_cv[data] = mi_cv

bio_channelsOPT_alphamu_real[experiment] = bio_channelsOPT_alphamu
bio_channelsOPT_alphamu_real[data] = mi_real

bio_channelsOPT_alphamu_cv[experiment] = bio_channelsOPT_alphamu
bio_channelsOPT_alphamu_cv[data] = mi_cv

bio_channelsPT_real[experiment] = bio_channelsPT
bio_channelsPT_real[data] = mi_real

bio_channelsPT_cv[experiment] = bio_channelsPT
bio_channelsPT_cv[data] = mi_cv

bio_channelsPT_alphamu_real[experiment] = bio_channelsPT_alphamu
bio_channelsPT_alphamu_real[data] = mi_real

bio_channelsPT_alphamu_cv[experiment] = bio_channelsPT_alphamu
bio_channelsPT_alphamu_cv[data] = mi_cv
```

```python
final = all_real.append(all_cv).reset_index().drop(columns = 'index')

final = final.append(all_channels_alphamu_real).reset_index().drop(columns = 'index')

final = final.append(all_channels_alphamu_cv).reset_index().drop(columns = 'index')

final = final.append(bio_channelsConly_alphamu_real).reset_index().drop(columns = 'index')

final = final.append(bio_channelsConly_alphamu_cv).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOConly_alphamu_real).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOConly_alphamu_cv).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOonly_alphamu_real).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOonly_alphamu_cv).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOPT_real).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOPT_cv).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOPT_alphamu_real).reset_index().drop(columns = 'index')

final = final.append(bio_channelsOPT_alphamu_cv).reset_index().drop(columns = 'index')

final = final.append(bio_channelsPT_real).reset_index().drop(columns = 'index')

final = final.append(bio_channelsPT_cv).reset_index().drop(columns = 'index')

final = final.append(bio_channelsPT_alphamu_real).reset_index().drop(columns = 'index')

final = final.append(bio_channelsPT_alphamu_cv).reset_index().drop(columns = 'index')
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.violinplot(x="experiment", y="test_accuracy", hue="data", data=final)#, split=True, rot = 90)
```
